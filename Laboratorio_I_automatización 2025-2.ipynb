**A**. **Definicion de** *def*:

Es una palabra clave usada para crear o definir una función, es decir, un bloque de código con nombre específico que almacena instrucciones para realizar una tarea determinada. Esta función puede ser llamada a lo largo del código para ejecutar dicho bloque.


```python
#Ejemplo: sumar dos números
a = int(input("Ingrese el primer número: "))
b = int(input("Ingrese el segundo número: "))
def suma(a, b):
  return a + b
resultado = suma(a, b)
print("El resultado de la suma es: ", resultado)
```

**C.** Desarrolle un algoritmo en el cual usted pueda definir el tamaño de una matriz cuadrada con
un numero N, posteriormente le pida cada uno de los valores de la matriz con la indicación
exacta de donde ira ubicado. Finalmente, el algoritmo debe retornar al usuario la matriz
deseada y su transpuesta.

```python
import numpy as np
n = int(input("Ingrese el tamaño de la matriz: "))
matriz = np.zeros((n,n))
for x in range(n):
  for y in range(n):
    print("valor para la posicion(",x+1,",",y+1,"): ",end="")
    matriz[x][y] =float(input())
print("matriz ingresada\n", matriz)
print("matriz transpuesta:\n", np.transpose(matriz))
```

D. Desarrolle un algoritmo que reciba un mes y día especifico, y nos dé una aproximación para esa
fecha sobre la estación climática en la cual estamos.

```python
#Se desarrollo con base en las fechas de las estaciones en el hemisferio Norte
def estacion(dia, mes):
  #Definir los meses y el número de días por mes
  meses = {
      "enero":31, "febrero" :28, "marzo":31, "abril":30, "mayo":31, "junio":30, "julio":31, "agosto":31, "septiembre":30, "octubre":31, "noviembre":30, "diciembre":31
  }
  #Clave: Nombre del mes
  #Valor: Cantidad de días del mes

  mes=mes.lower().strip()              #.lower va a convertir la entrada a minúsculas para que coincida con la clave del diccionario
  if mes not in meses:                 #.strip va a eliminar los espacios en blanco al principio y al final del texto ingresados por el usuario
    return "Mes inválido"

  if dia < 1 or dia > meses[mes]:
    return "Día inválido"

  #Definir las condiciones
  if (mes == "marzo" and dia >= 20) or (mes in ["abril","mayo"]) or (mes == "junio" and dia <= 20):
    estacion = "Primavera"
  elif (mes == "junio" and dia > 20) or (mes in ["julio", "agosto"]) or (mes == "septiembre" and dia <= 23):
    estacion = "verano"
  elif (mes == "septiembre" and dia > 23) or (mes in ["octubre", "noviembre"]) or (mes == "diciembre" and dia <= 22):
    estacion = "otoño"
  else:
    estacion = "Invierno"

  return (f"Para el día {dia} del mes de {mes} estaríamos en {estacion}")

# Obtener primero la entrada del mes, luego el día
mes = str(input("Ingrese el mes que desea consultar: "))
dia = int(input("Ingrese el día del mes que desea consultar: "))

print(estacion(dia, mes)) #Llamamos la función definida inicialmente
```

F. Una empresa de comercio electrónico necesita un programa en Python que procese reseñas de
clientes.
El programa debe:
1. Pedir al usuario que ingrese varias reseñas de productos (mínimo 3).
2. Convertir todas las reseñas a minúsculas y eliminar signos de puntuación básicos (. , ! ?).
3. Guardar cada reseña en una lista.
4. Crear un diccionario de frecuencias que indique cuántas veces aparece cada palabra en
todas las reseñas combinadas.
5. Identificar y mostrar:
Las 3 palabras más frecuentes en las reseñas.

```python
import string
reseñas = []
while True:
    num_res = int(input("Ingrese el número de reseñas: "))
    if num_res < 3:
        print("Error: El número de reseñas es inválido")
        continue  # Regresa al inicio
    for i in range (num_res):
      reseña = input("Ingrese la reseña: ")
      reseña = reseña.lower().translate(str.maketrans('', '', string.punctuation))
      reseñas.append(reseña)
    f = {}                  #Diccionario de frecuencias
    for reseña in reseñas:
      for palabra in reseña.split():  # Dividir cada reseña palabra por palabra
            if palabra in f:
                f[palabra] += 1
            else:
                f[palabra] = 1
    dic_asc = sorted(f.items(), key=lambda item: item[1], reverse = True)
    palabras_frec = dic_asc[:3]
    print("Las tres palabras más frecuentes en las reseñas son:\n" +
      "\n".join([f"{p}: {c}" for p, c in palabras_frec]))
    break
```
